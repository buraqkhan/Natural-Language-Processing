```python
import pandas as pd
import numpy as np
import nltk
import spacy
import random
from urduhack.normalization import normalize_characters
from urduhack.preprocessing import normalize_whitespace
from urduhack.preprocessing import remove_punctuation
from urduhack.preprocessing import remove_english_alphabets
unlp = spacy.blank('ur')
```

**Normalizing and Cleaning up the data**

```python
with open("faiz.txt") as f:
  lines = f.read()

lines = normalize_characters(lines)
lines = remove_punctuation(lines)
lines = remove_english_alphabets(lines)
lines = normalize_whitespace(lines)

start_words = []
end_words = []
corpus = lines.replace("\n", " NL ") # Adding ending and starting tags

text = unlp(corpus)
text = list(filter(None,  text))

for count, val in enumerate(text):
  if val.text == "NL":
    if count+1 < len(text):
      start_words.append(text[count+1].text)
    if val.text not in end_words:
      end_words.append(text[count-1].text)

print(end_words)
print(start_words)
```

# **Unigrams**

**Generating Language Model**

```python
unigrams = list()
for i in range(len(text)):
  if text[i].text == "NL":
    continue
  else:
    w1 = text[i].text
    unigrams.append(w1)

print(unigrams)
```

```python
uni_dict = {}

for x in unigrams:
  if x in uni_dict:
    uni_dict[x] += 1
  else:
    uni_dict[x] = 1

uni_dict = {k: v for k, v in sorted(uni_dict.items(), key=lambda item: item[1], reverse = True)}

for x in uni_dict:
  print(x, uni_dict[x])
```

**Generating Poem from Unigrams**

```python
def generateUnigramPoem():
  poem = ""
  for i in range(3):
    for j in range(4):
      line_length = random.randint(7, 10)
      
      rand_start = random.randrange(0, len(start_words)-1)
      rand_end = random.randrange(0, len(end_words)-1)

      starting = start_words[rand_start]
      ending = end_words[rand_end]

      line = starting
      for count in range(line_length):
        word = random.choice(unigrams) 
        line += " " + word
      
      poem += line + " " + ending +"\n" 
    poem += "\n"

  print(poem)    

generateUnigramPoem()
```

# **Bigrams**

**Generating Language Model**

```python
bigrams = list()

for i in range(len(text)-2):
  if text[i].text == "NL" or text[i+1].text == "NL" or text[i+2].text == "NL":
    continue
  else:
    w1 = text[i].text
    w2 = text[i+1].text
    bigrams.append(tuple((w1, w2)))

for x in bigrams:
  print(x[0], x[1])
```

```python
bi_dict = {}
bigram_dict = {}
for x in bigrams:
  if (x[0], x[1]) in bi_dict:
    bi_dict[x[0], x[1]] += 1
    bigram_dict[x[0], x[1]] += 1
  else:
    bi_dict[x[0], x[1]] = 1
    bigram_dict[x[0], x[1]] = 1
    
for x in bi_dict:
  bi_dict[x] = bi_dict[x] / uni_dict[x[0]]

bi_dict = {k: v for k, v in sorted(bi_dict.items(), key=lambda item: item[1], reverse = True)}

for x in bi_dict:
  print(x, bi_dict[x])
```

**Generating Bigram Poem**

```python
def generateBigrams():
  poem = ""
  for i in range(3):

    for j in range(4):
      line_length = random.randint(7, 10)
      
      rand_start = random.randrange(0, len(start_words)-1)
      rand_end = random.randrange(0, len(end_words)-1)

      starting = start_words[rand_start]
      ending = end_words[rand_end]

      line = starting
      max = []
      word = starting 
      
      for count in range(line_length-2):
        for val in bi_dict:
          if val[0] == word:
            max.append((val[0], val[1]))
        
        if len(max) == 0:
          word = random.choice(unigrams) 
          line += " " + word
        
        else:
          y = max[0]
          line += " " + y[1]
          word = y[1]
          max.clear() 

      poem += line + " " + ending +"\n" 
    poem += "\n"  

  print(poem)

generateBigrams()
```

# **Backward Bigrams**

**Generating Backward Bigram Poem using Bigram Language Model**

```python
def generateBackBigrams():
  poem = ""
  for i in range(3):

    for j in range(4):
      line_length = random.randint(7, 10)
      
      rand_start = random.randrange(0, len(start_words)-1)
      rand_end = random.randrange(0, len(end_words)-1)

      starting = start_words[rand_start]
      ending = end_words[rand_end]

      line = ending
      max = []
      word = ending
      
      for count in range(line_length-2):
        for val in bi_dict:
          if val[1] == word:
            max.append((val[0], val[1]))
        
        if len(max) == 0:
          word = random.choice(unigrams) 
          line += " " + word
        
        else:
          y = max[0]
          line += " " + y[0]
          word = y[0]
          max.clear() 

      # Reversing the string 
      line += " " + starting
      line = line.split(' ')
      line = ' '.join(reversed(line))

      poem += line + "\n" 
    poem += "\n"  

  print(poem)

generateBackBigrams()
```

Approach: Start at the end of a line, selecting a word from ending words and proceed to generate the line backwards till the number of words are equal to ones needed to be generated.

# **Bi-Directional Bigrams**

**Generating Bi-Directional Bigram Poem using Bigram Language Model**

```python
def generateBiDirectBigrams():
  poem = ""
  for i in range(3):

    for j in range(4):
      line_length = random.randint(7, 10)
      
      rand_start = random.randrange(0, len(start_words)-1)
      rand_end = random.randrange(0, len(end_words)-1)

      starting = start_words[rand_start]
      ending = end_words[rand_end]

      line = starting
      line_2 = ending
      max = []
      max_2 = []
      word = starting 
      word_2 = ending

      for count in range(int(line_length/2.5)):
        for val in bi_dict:
          if val[0] == word:
            max.append((val[0], val[1]))
          if val[1] == word_2:
            max_2.append((val[0], val[1]))
        
        if len(max) == 0:
          word = random.choice(unigrams) 
          line += " " + word
        else:
          y = max[0]
          line += " " + y[1]
          word = y[1]
          max.clear() 

        if len(max_2) == 0:
          word_2 = random.choice(unigrams)
          line_2 += " " + word_2
        else:
          y = max_2[0]
          line_2 += " " + y[0]
          word_2 = y[0]
          max_2.clear() 
        
      line_2 = line_2.split(' ')
      line_2 = ' '.join(reversed(line_2))

      poem += line + " " + line_2 +"\n" 
    poem += "\n"  

  print(poem)

generateBiDirectBigrams()
```

Approach:
Selecting a word from the list of starting words and placing it at the start of a line, and selecting a word from the list of ending words and placing it at the end of a line. 

I then use the logic for backward bigram on the ending word, and bigram on the first word until I've generated the number of words in a sentence that I need.
